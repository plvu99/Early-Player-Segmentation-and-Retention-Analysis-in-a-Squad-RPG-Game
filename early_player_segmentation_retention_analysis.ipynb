```python
# Read data
import pandas as pd
data = pd.read_csv('/content/drive/MyDrive/DSO 567/dso567_data_project1.csv')
data.head()
```

```python
import ast
# Parse all array columns into actual lists
array_cols = [
    'DX_TOTAL_SESSION_CT', 'DX_TOTAL_PLAYTIME', 'DX_TOTAL_SPEND',
    'DX_CAMPAIGN_MISSIONS_WON', 'DX_CAMPAIGN_MISSIONS_LOST', 'DX_CAMPAIGN_MISSIONS_RETREAT',
    'DX_PVP_MATCHES_WON', 'DX_PVP_MATCHES_LOST', 'DX_PVP_MATCHES_RETREAT',
    'DX_SIDE_MISSIONS_WON', 'DX_SIDE_MISSIONS_LOST', 'DX_SIDE_MISSIONS_RETREAT',
    'DX_START_ISIN_CHAPTER', 'DX_END_ISIN_CHAPTER'
]

days = ['D0','D1','D3','D7','D14','D21','D30','D60','D90','D120','D180','D240','D300','D365']

def parse_array(x):
    s = str(x).replace('undefined', '0').strip()
    return ast.literal_eval(s)

# Parse all array columns into list-of-lists
for col in array_cols:
    data[col + '_parsed'] = data[col].apply(parse_array)
```

```python
# Days index
import numpy as np
D0 = 0
D1 = 1

# Extract D0 and D1 for each key array
def extract_day(df, col, idx, label):
    arr_2d = np.vstack(df[col + '_parsed'].values)
    df[f'{col}_{label}'] = arr_2d[:, idx]

for col in array_cols:
    extract_day(data, col, D0, 'D0')
    extract_day(data, col, D1, 'D1')

# Show sample extracted columns
data[[
    'DX_TOTAL_SESSION_CT_D0','DX_TOTAL_SESSION_CT_D1',
    'DX_TOTAL_PLAYTIME_D0','DX_TOTAL_PLAYTIME_D1',
    'DX_TOTAL_SPEND_D0', 'DX_TOTAL_SPEND_D1'
]].head()
```

## Segmentation

### 1. Construct two segmentation groups

Early spenders are players who made any purchase on Day 0 or Day 1.

```python
# Define early spenders
data['Early_Spender'] = ((data['DX_TOTAL_SPEND_D0'] > 0) |
                         (data['DX_TOTAL_SPEND_D1'] > 0)).astype(int)
```

D1 retained players are those who returned to play on Day 1, having either at least one Day-1 session or any Day-1 playtime.

```python
# Define D1 retained players
data['D1_Retained'] = (
    (data['DX_TOTAL_SESSION_CT_D1'] > 0) |
    (data['DX_TOTAL_PLAYTIME_D1'] > 0)
).astype(int)
```

### 2. Extract Day-0 and Day-1 gameplay behavior

```python
# Compute total missions under each game mode
def sum_mode(df, prefix, day):
    return (
        df[f'{prefix}_WON_{day}'] +
        df[f'{prefix}_LOST_{day}'] +
        df[f'{prefix}_RETREAT_{day}']
    )

# Campaign Missions
data['D0_Campaign_Missions'] = sum_mode(data, 'DX_CAMPAIGN_MISSIONS', 'D0')
data['D1_Campaign_Missions'] = sum_mode(data, 'DX_CAMPAIGN_MISSIONS', 'D1')

# PVP Matches
data['D0_PVP_Matches'] = sum_mode(data, 'DX_PVP_MATCHES', 'D0')
data['D1_PVP_Matches'] = sum_mode(data, 'DX_PVP_MATCHES', 'D1')

# Side Missions
data['D0_Side_Missions'] = sum_mode(data, 'DX_SIDE_MISSIONS', 'D0')
data['D1_Side_Missions'] = sum_mode(data, 'DX_SIDE_MISSIONS', 'D1')
```

```python
# Compute total engagement
data['D0_Total_Missions'] = (data['D0_Campaign_Missions'] + data['D0_PVP_Matches'] + data['D0_Side_Missions'])
data['D1_Total_Missions'] = (data['D1_Campaign_Missions'] + data['D1_PVP_Matches'] + data['D1_Side_Missions'])
```

```python
# Compute session length or playtime quality
data['D1_Session_Length'] = data['DX_TOTAL_PLAYTIME_D1'] / data['DX_TOTAL_SESSION_CT_D1'].replace(0, 1)
```

### 3. Compute mean behavior for each segment

```python
# Summary report
seg_summary = data.groupby(['Early_Spender','D1_Retained'])[
    [
        'DX_TOTAL_SESSION_CT_D0','DX_TOTAL_PLAYTIME_D0',
        'D0_Total_Missions','D0_Campaign_Missions','D0_PVP_Matches','D0_Side_Missions',

        'DX_TOTAL_SESSION_CT_D1','DX_TOTAL_PLAYTIME_D1', 'D1_Session_Length',
        'D1_Total_Missions','D1_Campaign_Missions','D1_PVP_Matches','D1_Side_Missions'
    ]
].mean()

seg_counts = data.groupby(['Early_Spender','D1_Retained']).size()

print('Segment Counts:')
display(seg_counts)

print('\nSegment Behavior Summary:')
display(seg_summary)
```

**Meaning of the Segments**

**Segment 1 — Non-spender, Not Retained (0,0)**

Count: 31,787 players (~75%)

These players show very light Day-0 activity: 1.44 sessions, >17 minutes playtime, ~3.6 missions. They complete minimal onboarding content and do not return. This segment represents the typical early churn group.

**Segment 2 — Non-spender, D1 Retained (0,1)**

Count: 10,387 players (~25%)

These players show substantially deeper Day-0 behavior: 3.1 sessions, 87.5 minutes playtime, ~24 missions. They explore more modes, especially Campaign Missions.

Their Day-1 activity is also meaningful (3.6 sessions, ~80 minutes playtime). These are engaged free players who return despite not spending early.

**Segment 3 — Early Spender, Not Retained (1,0)**

Count: 20 players (<0.1%)

This segment shows high Day-0 engagement (3.1 sessions, 119 minutes playtime, 31 missions) but does not return on Day 1. This group reflects one-day spenders or tutorial-boost buyers.

**Segment 4 — Early Spender, D1 Retained (1,1)**

Count: 102 players

This is the most valuable segment. These players show the deepest early engagement: ~4.6 sessions on D0, ~218 minutes, >76 missions. There is heavy participation across game modes. Their engagement remains strong on Day 1 (7.6 sessions, ~212 minutes). Their behavior indicates they are highly invested, both in time and money, and are the core high-value users.

**Key Findings**

* D1 retained players played 2–5 times more Day-0 content than churners. This confirms that early gameplay depth is a major driver of short-term retention.

* Early spenders are extremely small (<0.3%) but 84% (102 players) retained. They exhibit the highest Day-0 and Day-1 activity. This segment is critical for long-term revenue.

* The largest segment is early churners (75%), who barely engage and have very low mission completion. This suggests onboarding friction or insufficient “first session excitement.”

* Returning non-spenders form a healthy engaged segment with strong Day-0 and Day-1 activity. This group represents high future monetization potential.

* D1 session length differentiates low-value vs. high-value retained players. Non-spender retained ~21 minutes per session, while spender retained ~33 minutes per session. Longer sessions signal deeper goal-oriented behavior.

## Clustering

### 1. K-Means clustering

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Define features for clustering (D0 and D1 engagement metrics)
features = [
    'DX_TOTAL_SESSION_CT_D0', 'DX_TOTAL_PLAYTIME_D0', 'D0_Total_Missions',
    'DX_TOTAL_SESSION_CT_D1', 'DX_TOTAL_PLAYTIME_D1', 'D1_Total_Missions',
    'D1_Session_Length', 'DX_TOTAL_SPEND_D0', 'DX_TOTAL_SPEND_D1'
]

# Handle potential infinite values resulting from division by zero in D1_Session_Length
# (though the previous step handled 0, it's good practice to check)
data.replace([np.inf, -np.inf], np.nan, inplace=True);
# Fill any remaining NaN values (e.g., from D1_Session_Length if DX_TOTAL_SESSION_CT_D1 was originally 0)
# We'll fill with 0, assuming no activity if it was undefined
data[features] = data[features].fillna(0);

X = data[features]

# Scale the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Determine optimal number of clusters using the elbow method (optional, but good for context)
# ssd = []
# K = range(1, 11)
# for k in K:
#     kmeans_model = KMeans(n_clusters=k, random_state=42, n_init=10)
#     kmeans_model.fit(X_scaled)
#     ssd.append(kmeans_model.inertia_)

# plt.figure(figsize=(10, 6))
# plt.plot(K, ssd, 'bx-')
# plt.xlabel('k')
# plt.ylabel('Sum of squared distances')
# plt.title('Elbow Method For Optimal k')
# plt.show()

# Based on the previous segmentation of 4 groups, let's start with 4 clusters
# or based on typical game analytics segments
k = 4 # Number of clusters
kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
clusters = kmeans.fit_predict(X_scaled)

data['Cluster'] = clusters

# Analyze cluster characteristics
cluster_summary = data.groupby('Cluster')[features].mean()
cluster_counts = data.groupby('Cluster').size().reset_index(name='Count')
cluster_counts['Percentage'] = (cluster_counts['Count'] / len(data)) * 100

print('Cluster Counts:')
display(cluster_counts)

print('\nCluster Behavior Summary (Mean values):')
display(cluster_summary)

# Visualize cluster sizes
plt.figure(figsize=(8, 6))
sns.barplot(x='Cluster', y='Count', data=cluster_counts)
plt.title('Distribution of Players Across Clusters')
plt.xlabel('Cluster')
plt.ylabel('Number of Players')
plt.show()

# Visualize average feature values per cluster
cluster_summary_unscaled = data.groupby('Cluster')[features].mean()
cluster_summary_unscaled.plot(kind='bar', subplots=True, layout=(3, 3), figsize=(15, 15), sharex=False, title='Average Feature Values per Cluster')
plt.tight_layout()
plt.show()
```

### 2. DBSCAN clustering

```python
from sklearn.cluster import DBSCAN

# DBSCAN clustering
# It is crucial to tune 'eps' and 'min_samples' for optimal results.
# A common starting point for min_samples is 2 * number of features.
# For eps, often determined by looking at a k-distance graph.

dbscan = DBSCAN(eps=0.5, min_samples=5)
# Using the same X_scaled data from KMeans
db_clusters = dbscan.fit_predict(X_scaled)

data['DBSCAN_Cluster'] = db_clusters

# Analyze DBSCAN cluster characteristics
db_cluster_summary = data.groupby('DBSCAN_Cluster')[features].mean()
db_cluster_counts = data.groupby('DBSCAN_Cluster').size().reset_index(name='Count')
db_cluster_counts['Percentage'] = (db_cluster_counts['Count'] / len(data)) * 100

print('DBSCAN Cluster Counts:')
display(db_cluster_counts)

print('\nDBSCAN Cluster Behavior Summary (Mean values):')
display(db_cluster_summary)

# Visualize DBSCAN cluster sizes
plt.figure(figsize=(8, 6))
sns.barplot(x='DBSCAN_Cluster', y='Count', data=db_cluster_counts)
plt.title('Distribution of Players Across DBSCAN Clusters')
plt.xlabel('DBSCAN Cluster (Noise points are -1)')
plt.ylabel('Number of Players')
plt.show()
```

**Better Performer: DBSCAN**

When comparing K-Means and DBSCAN on the D0 and D1 behavioral data, the two algorithms produced very different outcomes due to the heavily skewed nature of player activity that also included lots of outliers. K-Means performed poorly for this dataset, grouing 86% of players into one dominant cluster and creating a tiny cluster of 9 players driven entirely by outliers, which indicates that its assumption of spherical, evenly sized clusters does not fit real game behavior. This resulted in clusters that were imbalanced, difficult to interpret, and not actionable for segmentation.

In contrast, DBSCAN performed significantly better by adapting to the natural density structure of the data. It identified a meaningful set of clusters: casual players, light-moderate players, mid-core players. It also correctly isolated over 4,800 high-engagement outliers, including potential whales, as noise instead of forcing them into artificial centroids. DBSCAN also revealed several small but meaningful microsegments that K-Means could not detect. Because DBSCAN handles skew, non-linear cluster shapes, and extreme outliers while producing more interpretable and behaviorally distinct groups, it provides a significantly more accurate and useful clustering solution than K-Means in this scenario.
